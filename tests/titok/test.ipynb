```python
import sys
import os
sys.path.append(os.path.join(os.getcwd(), '..', '..', 'src'))
```

```python
from any_tokenizers.titok import TiTokImageTokenizer
```

```python

model_path = "yucornetto/tokenizer_titok_l32_imagenet"
tokenizer = TiTokImageTokenizer.from_pretrained(model_path)
```

```python
files = os.listdir("../images/")
files = [os.path.join("../images/", f) for f in files]
print(files)

tokens = tokenizer.encode_batch(files)
print(tokens)
```
